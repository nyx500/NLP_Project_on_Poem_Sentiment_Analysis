### Introduction to the Domain-Specific Area



Sentiment analysis entails the automated classification of texts based on their emotional tone. Sentiment analysis can be applied to a wide range of domains including "criminal communication detection" and "student feedback" (Ilmawan, Muladi, & Prasetya, 2024). This project, however, focuses on the multi-class text classification of lines of poetry (taken from the Google Research Datasets' poem-sentiment corpus) into different sentiment polarities or "semantic orientation" (Yusof, Mohamed, & Abdul-Rahman, 2015). Lines of English poetry from Project Gutenberg have been marked with one of four possible labels: 0 for "negative sentiment" or emotion, 1 for "positive sentiment," 2 for "neutral" or "no impact," and 3 for "mixed." This project aims to compare and analyse the performance of classical, statistical approaches to this multi-class classification problem with that of a deep learning model. Unlike typical sentiment analysis applications that detect **attitudes** towards entities, poetry often expresses various complex affective states, including moods, emotions and interpersonal stance (Jurafsky & Martin, 2024).

The vast availability of text data on the Web since the early 2000s has elevated the domain of sentiment analysis as a highly valuable research field. Liu (2012) outlines how the automated sentiment analysis using online customer reviews, forum posts, and social media comments is central to much current decision-making process in the business domain. It provides a more efficient cost-effective way of gauging customer satisfaction, especially when compared to traditional methods such as surveys and focus groups. Additionally, analysing attitudes in online posts can assist in predicting election results and sociopolitical trends. While the benefits of sentiment analysis for measuring customer satisfaction or political moods using online text data have been extensively documented, the rationale for sentiment analysis for poetic texts is not as immediately apparent as these clear financial or political advantages – however, there are several reasons it can be of value.

Poetic language constitutes a specific challenge for any text classification task due to the prevalence of figurative language and unconventional word usage (Kim & Klinger, 2019). Metaphorical and figurative expressions are also used in customer reviews and social media posts: thus, analysing and addressing the challenges raised during the computational detection of sentiment polarity in poetic language may improve the robustness of general sentiment detection techniques. Furthermore, Kim and Klinger argue that sentiment analysis techniques are of vital importance to the growing field of the "digital humanities" (Kim & Klinger, 2019). Sentiment analysis can shed important new perspectives on key research topics in literary studies, such as the evolution of the literary expression of affect over different historical periods or authorship attribution (Jurafsky & Martin, 2024). Importantly, sentiment analysis of literary texts can also be used to facilitate the detection of bias towards certain geographic locations or demographic groups over time. Sheng and Uthus, who compiled and annotated the dataset used in this project, did so in order to counteract the societal bias of an automated poetry collaboration tool (Sheng & Uthus, 2020). They trained a BERT-like deep-learning model to learn to identify the sentiment polarity in the lines of poetry randomly selected from Project Gutenberg, resulting in 85% accuracy.

As such, the field of sentiment analysis of poetic language can be valuable for refining NLP models' capability to handle figurative language, assisting in the formulation and testing of theories in literary research, and diagnosing social biases in seminal cultural texts.

## Handling the Missing "Mixed" Class Limitation

As shown here, only the train set contains (49) instances of class number 3 (mixed sentiment).

This raises an important issue which must be addressed before proceeding: what should be done in a situtation where **only** the training split contains "mixed" samples - particularly when considering that these "mixed" sentiment classes were associated with substantially less inter-annotator agreement? As Sheng and Uthus were interested in this dataset for a very particular use case (transforming lines of poetry expressing "negative" sentiment towards certain demographic groups into "positive" samples), the correct classification of poetic fragments of text expressing mixed or ambiguous emotion or mood may not have been the main priority of the authors. However, in the context where the primary objective *is* determining the capability of different models to detect sentiment in nuanced, figurative and literary language, being able to correctly identify "mixed" sentiment classes is pivotal to fully evaluating the robustness of existing text classification tools.

As such, substantial consideration was given to how to proceed and develop an adequate strategy for comparing the classification algorithms while accounting for this limitation. It was necessary to explore the possible ways to navigate around this shortcoming, and the advantages and disadvantages of each solution will now be clarified and summarized.

- **Adding more real samples**: First, the ideal solution would be to find more samples of poetic text expressing "mixed" (both negative and positive) sentiment. Unfortunately, although this method of enriching the existing dataset would increase the ability to truly evaluate the robustness of the models, it was regrettably not possible to find and employ two "expert validators" who would agree to do this given the time limitations and deadlines for this project.

- **Dataset "Augmentation"**: The HuggingFace user [AIManatee](https://huggingface.co/AiManatee) states that she "augmented" this dataset by adding 16 "mixed" sentiment samples to the validation and test sets and tested these new instances for ["semantic consistency, diversity (cosine similarity), length variation and novelty (ensuring the augmented data introduced new, relevant vocabulary)"](https://huggingface.co/AiManatee/RoBERTa_poem_sentiment), but unfortunately did not provide the updated, augmented dataset to the public (AIManatee, 2024). Augmenting the dataset in order to assess the RoBERTa model on the "mixed" class as well led to very minor decrease in accuracy from 0.8857142857142857 to 0.8842975206611571, and a decrease in F1-score from 0.9025108225108224 to 0.8813606847697756 (largely due to a decline in precision from 0.9201298701298701 to 0.8810538160090016) (AIManatee, 2024). This shows that when the "mixed" samples were included in the validation set, there were more false positives - although it would be better to be able to see in more detail which labels were misclassified more, this information was not provided.

- **Synthetic Sample Generation**: As described by Li et al. (2023), another option would be to generate synthetic text data using a large language model - however as stated by these authors "the effectiveness of the LLM-generated synthetic data in supporting
model training is inconsistent across different classification tasks" (Li et al., 2023). Silva, 2023).

- **Dataset Filtering**: An alternative would be simply to filter the dataset by removing the 49 "mixed" samples from the training data, like the original authors did (Sheng and Uthus, 2020), and to limit testing the ability of the classifiers to categorize the data under the remaining labels. Although this would be the easiest option, it is not ideal because mixed/ambiguous sentiment is something that is an important kind of affective language - in poetry and literature more generally, there are frequent expressions of emotions such as nostalgia, wistfulness, or melancholia (e.g. Thus, ignoring these samples would greatly diminish the importance and contribution made by this study.

- **Re-splitting the Dataset**: On could also combine the training, validation and test sets and create new splits that would contain the "mixed" class in the validation and test sets. However, changing the splits comes with the major disadvantage that the results could not be evaluated in context of the scores currently published online and would diminish the reproducibility and compatibility of this work.

Consequently, after some deliberation, it was decided that the classification methods used in this study will be applied *both* to the original train-dev-test split, *as well as* to a new, recombined version of the dataset which will ensure that samples from the "mixed" minority class are included in the validation and test sets. By performing this analysis on both configurations of the dataset, it may be possible to evaluate the impact of including more complex and ambiguous language in the dataset, while also ensuring comparability to existing achievements by running the experiments on the same dataset split.

### Evaluation Methodology

#### Training a Baseline Classifier
- Text strings will be pre-processed using basic methods (i.e. word tokenization, Bag-of-Words).
- A simple classifier (e.g. Multinomial Naive Bayes) will be trained on the training set to provide a benchmark for future experiments.
- This basic classifier will be assessed using the existing validation set. Along with accuracy, the precision, recall and F1 scores will be calculated for each of the three classes present in the original validation set.
- A confusion matrix will be used to visualize performance across classes.
-  It is vital to also calculate the recall and precision across the classes as well as "accuracy", as the dataset is heavily unbalanced.
-  Macro-averages for precision and recall will be computed. Macro-averaging tends to result in worse outcomes than micro-averaging because equal weight is given to each class (including the poorly-performing ones). Macro averages are useful when determining the performance for each class is "equally important”: here, we want to see how well the classifier performs over *all* the sentiment polarities (EvidentlyAI, 2024). Also, micro-averaging can "overemphasize the performance of the majority class", which can result in misleadingly "inflated" performance scores when the algorithm performs poorly on the less-represented categories.

#### Pre-Processing Experiments
- Several experiments will be run using the original split, e.g. adding negation handling, removing stopwords. 
- The same metrics as mentioned above will be calculated, tabulated and compared with baseline scores.
-  The highest-performing configuration will be used on the test set.

## Justification for using the Multinomial Naive Bayes Classifier as the Traditional Statistical Model

- A Multinomial Naive Bayes Classifier will be used as the traditional statisitcal model in this project.
-  The advantages of using the Naive Bayes classifier are that it is extremely fast and simple, has few tunable parameters, and can provide "a quick-and-dirty baseline for a classification problem" (Vanderplas, 2016).
-   Additionally, the *predict_proba* method enables one to view the posterior probability of each of the classes for a sample, which can provide an opportunity for a closer analysis of why certain samples were misclassified.
-    Furthermore, Naive Bayes tends to perform well when working with high-dimensional data (such as text data, where each sample's vector is often a long list of numbers denoting word counts), as clusters of instances in high dimensions tend to be more separate (Vanderplas, 2016).
- Nevertheless, as Jurafsky and Martin explain (Jurafsky & Martin, 2024), more sophisticated and highly-interpretable models such as logistic regression have important advantages over Naive Bayes.
    - When it calculates the likelihood of a sample having specific features given a particular class/label, Naive Bayes works on the assumption that the features are not correlated with each other, and assumes that every feature is independent.
    - However, in real life, features are seldom independent of one another. Consequently, logistic regression is more robust when there are many correlated features (Jurafsky & Martin, 2024).
    -  As a result, the authors conclude that when working with very large datasets *or* long documents (i.e. text samples), logistic regression tends to perform better, but that Naive Bayes can work just as well on short pieces of text.
    -   As the samples here consist of just one line of poetry, they definitely constitute very short text fragments. Wang & Manning have also shown that "for short snippet sentiment tasks, NB actually does better than
SVMs ["Support Vector Machines", another kind of classification algorithm] (while for longer documents the opposite result holds)" (Wang & Manning, 2012).
    - Additionally, Naive Bayes is much simpler and quicker to implement than Logistic Regression, as its performance does not depend on many iterations of tuning and testing model hyperparameters, such as learning rate, which is crucial when using other machine-learning algorithms.
    - Naive Bayes has been used successfully for decades for spam detection tasks, where datasets are usually extremely unbalanced (spam is a much rarer event than a legitimate email) (Jurafsky & Martin, 2024). In this particular case, the dataset is also very unbalanced, with the neutral class being much larger than any other category. As such, the advantages of using Naive Bayes for this task seem to outweigh the disadvantages: nonetheless, for future extensions to this project however, it would be interesting to compare results to a Logistic Regression classifier.
- Although support vector machines have been applied successfully in many text-classification scenarios(Sharma & Dey, 2012), and techniques exist for making these algorithms more robust, overall "the success of SVM is very limited when it is applied to the problem of learning from imbalanced datasets in which negative instances heavily outnumber the positive instances" (Akbani, Kwek & Japkowicz, 2004). Consequently, there are not many convincing arguments for why this model should be used as a starting point for analysis of this particular dataset.
- A decision tree is another kind of classifier that can be used for sentiment prediction tasks.
    - However, decision trees are very prone to overfitting (and thus performing poorly on unseen data) (Bramer, 2007).
    - Although there are different strategies for mitigating this risk, this involves extensive experimentation with hyperparameters such as configuring maximum tree depth or tree pruning settings, which introduces another layer of complexity to the task.
    -  Random Forest Classifiers have been applied successfully to sentiment classification tasks (see [this article](https://www.researchgate.net/publication/338050106_Sentiment_Analysis_Using_Random_Forest_Algorithm-Online_Social_Media_Based) on applying one to the Twitter sentiment dataset by Bahrawi, 2019) - however, this kind of ensemble classifier (which uses a collection of simple decision trees) has the disadvantage of being less interpretable.
- Maximum-Entropy classifiers can frequently perform better than Naive Bayes classifiers on certain datasets (Nigam, Lafferty, & McCallum, 1999), as it does not rely on low correlation between features - however, they take much longer to train and can overly rely on having a large set of training data for optimal performance (Vryniotis, 2013). ### Discussion of Techniques for Negation Handling in Sentiment Analysis

Negation handling is a valuable technique for improving the performance of sentiment classifiers. 

For instance, the sentence "it was a beautiful day" completely reverses the meaning of "it was not a beautiful day". As such, crafting features indicating negation can greatly impact the performance of classifiers applied to sentiment analysis. As Ilmawan, Muladi and Prasetia show (2024), at the moment there are myriad techniques that one can experiment with to signal the presence of negation within text, including both symbolic and data-driven methods. These authors note that although sentiment classification might seem straightforward, it actually poses many challenges for a range of NLP tasks. 

One of these most substantial challenges is negation, which these authors argue can be seen as "the most crucial challenge in 
sentiment analysis and has the most significant impact" (Ilmawan, Muladi, & Prasetya, 2024) Negation handling or "negation scope detection" is further complicated by the fact that negation can be explicit, as signalled by cues such as the word "not", as well as implicit: "it was the first and last time he ever did a good job". Rule-based approaches to solving this challenge include marking the rest of the sentence following the negation cue as opposite polarity, or at least marking the reset of the words until the next punctuation symbol as such, as well as more complex approaches which only mark words associated with specific POS-tags as negated or dependency parsing to determine grammatical contexts. Different machine-learning techniques for negation detection such as SVMs, deep-learning networks like BiLSTMs and transformer-based models have also shown a lot of promise (Ilmawan, Muladi, & Prasetya, 2024). Advanced hybrid techniques such as PSDEE (polarity shift detection, elimination, and ensemble) combine rule-based and statistical approaches for superior performance. Rule-based approaches are usually faster to apply (circumventing the need for training models) and work well on simple, short sentences, while new machine-learning techniques (considered SOTA) can capture more complex relationships. 

For this study, a simple feature augmentation function will be defined to mark all words following a negation cue (e.g. "not" or "never") up until the next punctuation mark as negative, as detailed in Jurafsky and Martin (2024). However, one should bear in mind that this simple approach might not work in all cases (i.e. the word "not" does not always indicate inversion of sentiment, for instance in the phrase "not only does he cook but he also cleans"). It is useful to note that more complex and higher-performing techniques exist, which might be potential avenues to explore for future research. Another technique that will be experimented with in this project is outlined [in this algorithm written by Utkarsh Lal (2022)](https://towardsdatascience.com/increasing-accuracy-of-sentiment-classification-using-negation-handling-9ed6dca91f53). Lal writes that this algorithm for negation feature engineering can even make simple classifiers like Naive Bayes or Logistic Regression perform as well as complex embedding-based models. This approach uses the WordNet library to exchange the word following a negation cue (e.g. "not") to its antonym (e.g. changing "not pretty" to "ugly"). If a word does not have any antonym, it suggests it is a verb or word does not have an opposite available, and the sentence is left as it is. The antonym with the highest dissimilarity (calculated using the *wup_similarity*) method "ensures that the maximum inversion of polarity takes place" (e.g. mitigating the risk of exchanging expressions such as "not beautiful" to "bad"). This article visualizes the dramatic increase in accuracy using logistic regression once this negation algorithm was applied. As the context is left unchanged if a word does not have an antonym in WordNet, this slightly more advanced approach can hopefully reduce the risk of negating the following words when an apparent negation cue does not actually signal negation.

### Feature Engineering Experiment #1 for Statistical Classifier with Original Dataset Split: Negation Handling using _NOT Prefixes Following a Negation Cue

In this section, features will be added indicating that polarity of a fragment of text is negative following a "negation word" (e.g. not, nor, never) using the method detailed in Jurafsky and Martin (2024). For each term following a negation word, a NOT_ will be used to prefix each following word.

```python
# Define negation patterns
negation_patterns = [
    r"\bnot\b",     # Matches the whole word "not"
    r"not$",        # Match "not" at the end of the word, e.g. "cannot"
    r"\bno\b",      # Match the whole word "no"
    r"n't$",        # Match "n't" at the end of the word, e.g. "shouldn't"
    r"\bnever\b",
    r"\bnor\b",
    r"\bnone\b",
    r"\bnothing\b",
    r"\bnowhere\b",
    r"\bneither\b"
]

# List of punctuation markers now including "..."
punctuation_markers = list(string.punctuation)
punctuation_markers.append("...")

def simple_negation_features(tokens):
    add_negation = False  # Flag to indicate whether to add "_NOT"
    negation_prefix = "NOT_"  # Prefix to prepend to tokens indicating negation
    negated_tokens = []  # List to store tokens with negation markers
    # Iterate over each token
    for token in tokens:
        # Check if the token matches any of the negation patterns, thus is a negation cue
        if any(re.match(cue, token.lower()) for cue in negation_patterns):
            add_negation = True  # Set the flag to add "_NOT" to subsequent tokens
            negated_tokens.append(token)  # Add current token as-is to negated_tokens
            continue # proceed to the next token
            
        # If this token is a punctuation marker then reset the flag to stop adding "_NOT" as a prefix
        if token in punctuation_markers:
            add_negation = False
            
        # Add the "_NOT" prefix to tokens when add_negation flag is on
        if add_negation:
            negated_tokens.append(negation_prefix + token)  # Add "_NOT" prefix to token
        else:
            negated_tokens.append(token)  # Add the original token if no negation or if it's punctuation
    return negated_tokens
```

### Feature Engineering Experiment #2 for Statistical Classifier with Original Dataset Split: Negation Handling using WordNet

Following the alternate method for negation handling detailed [here](https://towardsdatascience.com/increasing-accuracy-of-sentiment-classification-using-negation-handling-9ed6dca91f53) uses WordNet to generate synsets (sets of "synonymous words that express the same concept") for the word imemdiately following a negation cue such as "not" or "never". After generating the synsets, the algorithm checks if WordNet contains an antonym (word with the opposite meaning) for this word. The word is left as it is if there is no available antonym: thus, this approach is more sophisticated than the rudimentary algorithm constructed above for negation handling. For instance, verbs do not have antonyms, so expressions such as "he did not envy" will not be negated, allowing for a more nuanced strategy. This new algorithm selects the antonym with the highest dissimilarity score (calculated as *dissimilarity = (1 — word1.wup_similarity(word2))*) to ensure that the maximum polarity reversal takes place. The negation-cue word preceding the antonym is then removed.

```python
## CODE ADAPTED FROM https://towardsdatascience.com/increasing-accuracy-of-sentiment-classification-using-negation-handling-9ed6dca91f53 ###########

# define a function for this more sophisticated type of negation handling.

"""
    Reverses the meaning of tokens following a negation cue to its antonym (if one exists) using WordNet synsets.
    Originally, the algorithm only replaced words following negation cues with their antonym if they immediately followed
    a negation cue (e.g. "she was not beautiful" --> "she was ugly"). However, it was unable to reverse the meaning of sentiment-carrying words
    that appeared in a fragment of text where there was a determiner (e.g. "the", "a") between the "negation cue" word and the 
    word that should be inverted (e.g. "it was not A nice day" --> "it was an ugly day").
"""
def handle_negation_with_wordnet(tokens, negation_patterns):	
  """
  Input: Tokenized sentence (List of words)
  Output: Tokenized sentence with negation handled (List of words)
  """
  temp = int(0)
  pos_tags = pos_tag(tokens)  # Get POS tags for the tokens to handle cases where negation cue is folowed by a determiner
  for i in range(len(tokens)):
      # if sentence[i-1] in ['not',"n't"]: --> I changed this line of code to handle more complex regex negation patterns I defined before
      if any(re.search(pattern, tokens[i-1]) for pattern in negation_patterns): # checks if the last token is a negation cue using the regex I defined.

          ######################################## MY ADAPTATION HERE ###############################################################################
          # If the current token is a determiner and next token is an adverb (e.g. very, most, quite) then target the word following the adverb
          if pos_tags[i][1] in ['DT'] and i + 2 < len(tokens) and pos_tags[i + 1][1] in ['RB']:
              index_of_word_to_replace = i + 2
         # If the current token is a determiner and next word is not an adverb then target the word immediately after the determiner
          elif pos_tags[i][1] in ['DT'] and i + 1 < len(tokens):
              index_of_word_to_replace = i + 1
          # Current token is not a determiner, so target the current token
          else:
              index_of_word_to_replace = i

          ###########################################################################################################################################
          
          antonyms = []
          # Iterates over the synsets for the current token.
          for syn in wordnet.synsets(tokens[index_of_word_to_replace]):
              syns = wordnet.synsets(tokens[index_of_word_to_replace])
              # get the name of the first synset for this token from wordnet
              w1 = syns[0].name()
              # reset temp (stores dissimilarity score) for each iteration of the loop
              temp = 0
              # go through the synonyms/lemmas (dictionary forms of words) in the synset
              for l in syn.lemmas():
                  # if the lemma has antonyms, append them to the antonyms list initiated as an empty list in each iteration of the tokens-loop.
                  if l.antonyms():
                      antonyms.append(l.antonyms()[0].name()) # append name of the antonym to the list
              max_dissimilarity = 0
              # now iterate over each antonym for this particular token (if it has one) and calculate the wup similiarity score
              # between the current token synset and the antonym synset.
              for ant in antonyms:
                  syns = wordnet.synsets(ant)
                  w2 = syns[0].name()
                  syns = wordnet.synsets(tokens[index_of_word_to_replace])
                  w1 = syns[0].name()
                  word1 = wordnet.synset(w1)
                  word2 = wordnet.synset(w2)
                  # wup = Wu-Palmer similarity. It measures similarity of 2 words using depth of their wordnet common ancestor.
                  if isinstance(word1.wup_similarity(word2), float) or isinstance(word1.wup_similarity(word2), int): # check sim score is float or int.
                      # use the 'temp' variable to store current dissimilarity score.
                      temp = 1 - word1.wup_similarity(word2) # calculate dissimilarity score by subtracting similarity score from 1.
                  # if new dissimilarity score ('temp') exceeds 'max_dissimilarity' (initialized to 0), update max_dissimilarity.
                  if temp > max_dissimilarity:
                      max_dissimilarity = temp
                      # replace current token with max-dissimilarity antonym
                      antonym_max = ant
                      tokens[index_of_word_to_replace] = antonym_max
                      # removes the negation cue (e.g. "not")
                      tokens[i-1] = ''

                  # If the determiner is "a" or "an", and an antonym has been foudn, modify determiner to adapt to first letter of next word
                  if tokens[index_of_word_to_replace-1] == "a" and tokens[index_of_word_to_replace][0].lower() in 'aeiou': # antonym starts with a vowel
                      tokens[index_of_word_to_replace-1] = "an"
                  elif tokens[index_of_word_to_replace-1] == "an" and tokens[index_of_word_to_replace][0].lower() not in 'aeiou': # antonym starts with a consonant
                      tokens[index_of_word_to_replace-1] = "a"
        
  # remove the '' markers used to delete negation words after antonym was used
  while '' in tokens:
      tokens.remove('')
  return tokens

######################################################################################################################################################
```

### Feature Engineering Experiment #6 for Statistical Classifier with Original Dataset Split: Negation Handling using <> and Lemmatization

Lemmatization involves converting tokens into their dictionary/root form (lemmas). For example, "runs" will be contracted to "run". Sometimes, this can reduce token variability in the number of tokens and can thus assist the classifier in learning new patterns by making it clearer which words are similar. Therefore, we will attempt this experiment in the below code cell.

```python
############################## REFERENCE: CODE ADAPTED FROM https://medium.com/@yashj302/lemmatization-f134b3089429 ###########################

def get_wordnet_pos(token):
    """A helper function: maps POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([token])[0][1][0].lower()
    # Convert nltk.pos_tag to a tag accepted by WordNetLemmatizer
    tag_dict = {"a": wordnet.ADJ,
                "n": wordnet.NOUN,
                "v": wordnet.VERB,
                "r": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_tokens(lemmatizer, tokens):
    """
    Lemmatizes a list of tokens.

    Inputs:
    lemmatizer (instance of a lemmatizer), tokens = a list of tokens

    Outputs:
    a list of lemmatized tokens
    """
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]
    return lemmatized_tokens
    
##################################################################################################################################################        
```

### Feature Engineering Experiment #7 for Statistical Classifier with Original Dataset Split: Negation Handling using WordNet and AFINN Sentiment Lexicon

Because the experiments using standard text-processing techniques so far have failed to increase the average F1-score beyond 54%, I will now attempt to construct a hybrid approach by combining Multinomial Naive Bayes with features generated using a sentiment lexicon (as inspired by [this article](https://download.garuda.kemdikbud.go.id/article.php?article=1304227&val=146&title=Hybrid%20approach%20naive%20bayes%20and%20sentiment%20VADER%20for%20analyzing%20sentiment%20of%20mobile%20unboxing%20video%20comments) (Chaithra, 2019), where Naive Bayes is combined with the sentiment VADER lexicon). Jurafsky and Martin list a wide range of possible sentiment lexicons (manually-curated lists of positive and negative words), such as the General Inquirer, MPQA Subjectivity Lexicon (for detecting objective/subjective conntations of words), and Hu and Liu's polarity lexicon (Jurafsky & Martin, 2024).  

For this task, I will use the AFINN polarity lexicon because of its "simplicity" [(rakshita_iyer, 2023)](https://www.geeksforgeeks.org/python-sentiment-analysis-using-affin/) and easy integrability with the Python environment.

After downloading the lexicon, as well as adding word features indicating the presence of a word, I will also add features for sentiment score for each word calculated using the AFINN library.

In AFINN, a score of 0.0 indicates neutral sentiment, a negative score indicates negative sentiment, and a positive score indicates positive sentiment. However, some limitations of this lexicon must be noted: as it outputs scores for individual words, the word "terribly" in the context of "they were terribly happy" will lead to a negative score being assigned to "terribly", even though the meaning here is to strengthen the positive affect expressed by "happy". Instead, the negative score for "terribly" will cancel out the positive score for "happy", leading to a neutral score. These kind of juxtapositions of typically negative modifiers to strengthen the meaning of words with positive connotations (and vice versa) are frequently found in poetic and literary language, which also includes other unusual uses of words in novel contexts. As a result, sentiment lexicons might not be sufficient to achieve reasonably high accuracy and f1-scores on poetic texts - highlighting the core limitations of using word-counting and feature-engineering approaches with traditional statistical classifiers on this particular kind of text. When applied to product reviews or political opinions (which typically use more direct, typical and straightforward language), these approaches can be useful, but it seems that the dependence of words on the *context* of the whole document is more imperative when handling literary texts: thus potentially justifying the use of deep-learning and transformer-based methods for this domain of text documents.

Vader is not another frequently-used sentiment lexicon, but its disadvantage for this project is that it is ["specifically attuned to sentiments expressed in social media"](https://github.com/cjhutto/vaderSentiment) (Hutto, n.d.), rather than for literary texts.

### Feature Engineering Experiment #8 for Statistical Classifier with Original Dataset Split: Negation Handling using WordNet and SentiWordNet Sentiment Lexicon

```python
########## CODE ADAPTED FROM: https://medium.com/@nidhikayadav/sentiment-analysis-with-python-sentiwordnet-fd07ffc557 ##############################
def get_sentiwordnet_scores_from_tokens(tokens):

    # remove bigrams to calculate sentiment score for the document (tokenized line of poetry
    tokens_without_bigrams = [token for token in tokens if not isinstance(token, tuple)]        
    
    pos_tagged_tokens = nltk.pos_tag(tokens_without_bigrams)
    
    positive_sentiment = 0.0
    negative_sentiment = 0.0
    
    for word_tag_pair in pos_tagged_tokens:
        word = word_tag_pair[0]
        pos_tag = word_tag_pair[1]
        # Convert nltk pos_tags to WordNet format for the most relevant pos tags
        if pos_tag.startswith('J'):
            pos_tag =  wordnet.ADJ
        elif pos_tag.startswith('R'):
            pos_tag =  wordnet.ADV    
        elif pos_tag.startswith('N'):
            pos_tag =  wordnet.NOUN
        else:
            continue
        word_synsets = wordnet.synsets(word, pos=pos_tag)
        if not word_synsets:
            continue
        top_synset = word_synsets[0]
        senti_word_net = swn.senti_synset(top_synset.name())
        positive_sentiment += senti_word_net.pos_score() 
        negative_sentiment +=  senti_word_net.neg_score()
    return positive_sentiment, negative_sentiment

####################################################################################################################################################
```

## Performance of Multinomial Naive Bayes Classifier on Original Dataset Split with WordNet Negation Handling and SentiWordNet Scores

The performance of this classifier on the test set was very poor, as the macro-average F1-score was only 48%, and the confusion matrix shows that hardly any positive or negative samples were correctly classified.

To a large extent, this is probably due to the small size of the validation and test sets (only about 100 samples) and the disproportionally small number of positive and negative samples compared to the dominant "neutral" class. The classifier probably had insufficient data to learn from.

Furthermore, the traditional approach to NLP text classification does not take into account the context of the words, and is therefore limited in its ability to infer patterns from small sets of samples, compared to a neural network that can capture relationships in a sequence of words.

As a result, I will now repeat the above experiments using five-fold stratified cross-validation that ensures that each class (including the "mixed" class) is proportionally represented in each fold, as well as in the training and test splits. One reason for this decision is that using cross-validation uses all of the training data to evaluate the best settings and features, which is vital when the dataset is so small and there are limited samples of the minority classes. Furthermore, this enables evaluation on the "mixed" sentiment class which was missing from the validation and test sets in the original dataset splits.
